```
import matplotlib.pyplot as plt
import pandas as pd
```

```
# Reading a csv file
file=pd.read_csv("student.csv")
print(file)
# Adding title to pie chart
plt.title("Student Marks Comparison")
# Here value will be marks and labels will be name of students
plt.pie(file.iloc[0:,1],labels=file.iloc[0:,0],autopct='%.1f%%',shadow=True) 
plt.show()
plt.title("Student Study_hours Comparison")
# Here values will be study_hours and labels again be name of students
plt.pie(file.iloc[0:,-1],labels=file.iloc[0:,0],autopct='%.1f%%',shadow=True)
plt.show()
```
