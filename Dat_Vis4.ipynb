```
from bs4 import BeautifulSoup 
import requests
```

```
url="https://en.wikipedia.org/wiki/Machine_learning"
response=requests.get(url)
```

```
data=response.text
soup=BeautifulSoup(data,'lxml')
```

```
a=[]
for i in soup.find_all('p'):
  for j in i.text.split(" "): 
    a.append(j)
di={}
for i in a:
  di.update({i:a.count(i)})
```
